## Project Goal
This notebook serves as the central integration point for the FX Decision Recommendation System. It pulls outputs from the team's data preprocessing, risk scoring, and forecasting modules to provide a final FX conversion recommendation for Indian businesses.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# Set paths
DATA_PATH = '../data/processed/cleaned_fx_data.csv'
print("Setup Complete. Ready for Integration.")
```

## 1. Data Integration (Srividya's Output)
Loading the preprocessed USD-INR exchange rate data.

```python
if os.path.exists(DATA_PATH):
    df = pd.read_csv(DATA_PATH, index_col=0, parse_dates=True)
    print("Successfully loaded cleaned FX data.")
    display(df.tail())
else:
    print("Warning: Cleaned data file not found at", DATA_PATH)
```

## 2. Risk Scoring Integration (Kanishkhan's Task)
Placeholder for the Risk Scoring engine. High volatility indicates high risk.

```python
def get_risk_status(volatility):
    """
    Placeholder logic for Risk Scoring.
    To be updated by Kanishkhan.
    """
    threshold = 0.005 # Example threshold
    if volatility > threshold:
        return "HIGH RISK"
    return "LOW/STABLE RISK"

latest_volatility = df['USD_Volatility'].iloc[-1] if 'USD_Volatility' in df.columns else 0.006
risk_status = get_risk_status(latest_volatility)
print(f"Current Risk Status: {risk_status} (Volatility: {latest_volatility:.4f})")
```

## 3. Forecasting Baseline (Adwaitha's Task)
Visualization of the current trend.

```python
plt.figure(figsize=(10, 5))
df['USD'].tail(30).plot(title="USD-INR Last 30 Days Trend")
plt.ylabel("Exchange Rate (INR / 1 USD)")
plt.grid(True)
plt.show()

predicted_trend = "UP" # Example forecast output
print(f"Forecasted Movement: {predicted_trend}")
```

## 4. Scenario Impact (Aadhithya's Task)
Calculating 'Money at Stake'.

```python
def calculate_impact(amount_usd, rate_change):
    impact_inr = amount_usd * rate_change
    return impact_inr / 100000 # Return in Lakhs

payment_amount = 100000 # $100,000
daily_drop = 0.50 # INR drop
loss_lakhs = calculate_impact(payment_amount, daily_drop)

print(f"Business Impact: A {daily_drop} INR movement on ${payment_amount:,} results in a {loss_lakhs:.2f} Lakh INR impact.")
```

## 5. Decision Recommendation Engine (Adarsh's Task)
Combining Risk and Forecast to provide actionable advice.

```python
def generate_recommendation(risk, forecast_trend):
    if risk == "HIGH RISK":
        if forecast_trend == "UP":
            return "CONVERT 100% IMMEDIATELY (Lock in rates now)"
        else:
            return "CONVERT 50% (Hedge against volatility)"
    else:
        if forecast_trend == "UP":
            return "WAIT / CONVERT PARTIALLY (Market is trending up)"
        else:
            return "DELAY CONVERSION (Stable market)"

advice = generate_recommendation(risk_status, predicted_trend)

print("--- FX DECISION REPORT ---")
print(f"Risk Status: {risk_status}")
print(f"Market Trend: {predicted_trend}")
print(f"RECOMMENDATION: {advice}")
```
