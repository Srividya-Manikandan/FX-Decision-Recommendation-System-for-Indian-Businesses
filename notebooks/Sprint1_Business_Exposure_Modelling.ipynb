## Scenario-Based Business Exposure Analysis


## Assumptions

The following assumptions are used for business exposure modelling:

- USD exposure per transaction: $10,000  
- Base USD–INR exchange rate: 83  
- IT Services Firm INR operational costs: ₹6,00,000  
- Scenarios are defined to represent favorable, neutral, and adverse currency movements  

All values are assumed for academic demonstration and sensitivity analysis purposes.


```python
import pandas as pd

# Base assumptions
usd_amount = 10000
base_rate = 83
it_costs = 600000

usd_amount, base_rate, it_costs
```

## Scenario-Based Exposure Analysis

To move beyond single-point estimation, business exposure is evaluated under
multiple realistic exchange rate scenarios.


```python
# Define scenarios
scenarios = pd.DataFrame({
    "Scenario": ["Favorable", "Neutral", "Adverse"],
    "USD_INR_Rate": [82, 83, 85]
})

# Importer exposure (loss when rate increases)
scenarios["Importer_Impact_INR"] = usd_amount * (scenarios["USD_INR_Rate"] - base_rate)

# Exporter exposure (loss when rate decreases)
scenarios["Exporter_Impact_INR"] = usd_amount * (base_rate - scenarios["USD_INR_Rate"])

# IT firm profit
scenarios["IT_Firm_Profit_INR"] = (usd_amount * scenarios["USD_INR_Rate"]) - it_costs

scenarios
```

## Profit-at-Risk (PaR) Analysis

Profit-at-Risk represents the maximum expected loss faced by a business
under an adverse exchange rate movement.


```python
# Identify adverse scenario
adverse_rate = scenarios.loc[scenarios["Scenario"] == "Adverse", "USD_INR_Rate"].values[0]

profit_at_risk = pd.DataFrame({
    "Business Type": ["Importer", "Exporter", "IT Services Firm"],
    "Profit_at_Risk_INR": [
        usd_amount * (adverse_rate - base_rate),          # Importer loss
        usd_amount * (base_rate - adverse_rate),          # Exporter loss
        (usd_amount * base_rate) - (usd_amount * adverse_rate)  # IT firm downside
    ]
})

profit_at_risk
```

## Comparative Sensitivity Analysis

This analysis compares how sensitive different business types are to a
₹1 movement in the USD–INR exchange rate.


```python
# Sensitivity per ₹1 movement
sensitivity = pd.DataFrame({
    "Business Type": ["Importer", "Exporter", "IT Services Firm"],
    "Impact_per_1_INR_Move": [
        usd_amount,            # Importer
        usd_amount,            # Exporter
        usd_amount             # IT firm revenue sensitivity
    ]
})

# Add qualitative risk levels
sensitivity["Risk_Level"] = ["High", "Medium", "High"]

sensitivity
```

## Business Interpretation & Decision Implications

### Importer
Importers are highly exposed to INR depreciation. Even a ₹1 increase in the
USD–INR rate results in a ₹10,000 increase in costs. Under adverse scenarios,
importers face significant profit-at-risk, indicating the need for early
currency conversion or hedging strategies.

### Exporter
Exporters benefit from INR depreciation but face downside risk when the INR
strengthens. While exporters can gain in favorable scenarios, adverse movements
can reduce INR realizations, making timing decisions critical.

### IT Services Firm
IT services firms are structurally sensitive to exchange rate movements due to
USD-denominated revenues and INR-denominated costs. Although revenue increases
with higher USD–INR rates, firms with thin margins face substantial downside risk
if the currency moves unfavorably.

### Summary
The analysis highlights that exchange rate exposure is asymmetric across
business types. Importers and IT firms exhibit high sensitivity, while exporters
face conditional risk depending on currency direction. These insights form a
critical input for risk scoring and FX decision recommendation modules in
subsequent sprints.
