## Scenario-Based Business Exposure Analysis


## Assumptions

The following assumptions are used for business exposure modelling:

- USD exposure per transaction: $10,000  
- Base USD–INR exchange rate: 83  
- IT Services Firm INR operational costs: ₹6,00,000  
- Scenarios are defined to represent favorable, neutral, and adverse currency movements  

All values are assumed for academic demonstration and sensitivity analysis purposes.


```python
import pandas as pd

# Base assumptions
usd_amount = 10000
base_rate = 83
it_costs = 600000

usd_amount, base_rate, it_costs
```

## Scenario-Based Exposure Analysis

To move beyond single-point estimation, business exposure is evaluated under
multiple realistic exchange rate scenarios.


```python
# Define scenarios
scenarios = pd.DataFrame({
    "Scenario": ["Favorable", "Neutral", "Adverse"],
    "USD_INR_Rate": [82, 83, 85]
})

# Importer exposure (loss when rate increases)
scenarios["Importer_Impact_INR"] = usd_amount * (scenarios["USD_INR_Rate"] - base_rate)

# Exporter exposure (loss when rate decreases)
scenarios["Exporter_Impact_INR"] = usd_amount * (base_rate - scenarios["USD_INR_Rate"])

# IT firm profit
scenarios["IT_Firm_Profit_INR"] = (usd_amount * scenarios["USD_INR_Rate"]) - it_costs

scenarios
```

## Profit-at-Risk (PaR) Analysis

Profit-at-Risk represents the maximum expected loss faced by a business
under an adverse exchange rate movement.


```python
# Identify adverse scenario
adverse_rate = scenarios.loc[scenarios["Scenario"] == "Adverse", "USD_INR_Rate"].values[0]

profit_at_risk = pd.DataFrame({
    "Business Type": ["Importer", "Exporter", "IT Services Firm"],
    "Profit_at_Risk_INR": [
        usd_amount * (adverse_rate - base_rate),          # Importer loss
        usd_amount * (base_rate - adverse_rate),          # Exporter loss
        (usd_amount * base_rate) - (usd_amount * adverse_rate)  # IT firm downside
    ]
})

profit_at_risk
```

## Comparative Sensitivity Analysis

This analysis compares how sensitive different business types are to a
₹1 movement in the USD–INR exchange rate.


```python
# Sensitivity per ₹1 movement
sensitivity = pd.DataFrame({
    "Business Type": ["Importer", "Exporter", "IT Services Firm"],
    "Impact_per_1_INR_Move": [
        usd_amount,            # Importer
        usd_amount,            # Exporter
        usd_amount             # IT firm revenue sensitivity
    ]
})

# Add qualitative risk levels
sensitivity["Risk_Level"] = ["High", "Medium", "High"]

sensitivity
```

## Business Interpretation & Decision Implications

### Importer
Importers are highly exposed to INR depreciation. Even a ₹1 increase in the
USD–INR rate results in a ₹10,000 increase in costs. Under adverse scenarios,
importers face significant profit-at-risk, indicating the need for early
currency conversion or hedging strategies.

### Exporter
Exporters benefit from INR depreciation but face downside risk when the INR
strengthens. While exporters can gain in favorable scenarios, adverse movements
can reduce INR realizations, making timing decisions critical.

### IT Services Firm
IT services firms are structurally sensitive to exchange rate movements due to
USD-denominated revenues and INR-denominated costs. Although revenue increases
with higher USD–INR rates, firms with thin margins face substantial downside risk
if the currency moves unfavorably.

### Summary
The analysis highlights that exchange rate exposure is asymmetric across
business types. Importers and IT firms exhibit high sensitivity, while exporters
face conditional risk depending on currency direction. These insights form a
critical input for risk scoring and FX decision recommendation modules in
subsequent sprints.

## Break-Even and Pain-Threshold Analysis

While scenario-based exposure quantifies potential gains or losses, businesses
also require decision boundaries to determine when exchange rate movements
become unacceptable.

- **Break-even rate**: Exchange rate at which profit impact is zero
- **Pain-threshold rate**: Exchange rate beyond which losses are considered
  operationally unacceptable

These thresholds help convert exposure values into actionable business signals.


```python
# Break-even and pain-threshold definitions
thresholds = pd.DataFrame({
    "Business": ["Importer", "Exporter", "IT Services Firm"],
    "Break_Even_Rate": [83, 83, 83],
    "Pain_Threshold_Rate": [85, 81, 82]
})

thresholds
```

## Exposure Zone Classification

To simplify decision-making, exchange rate outcomes are classified into
qualitative exposure zones:

- **Safe**: No immediate business action required
- **Warning**: Monitoring or partial intervention required
- **Danger**: Immediate action required to mitigate losses

This classification supports intuitive business interpretation without
introducing automated decision rules.


```python
# Exposure zone classification for Importer
def importer_exposure_zone(rate):
    if rate <= 83:
        return "Safe"
    elif rate < 85:
        return "Warning"
    else:
        return "Danger"

scenarios["Importer_Exposure_Zone"] = scenarios["USD_INR_Rate"].apply(importer_exposure_zone)

scenarios
```

## Multi-Currency Exposure Extension

The exposure modelling framework is extended to support multiple foreign
currencies commonly used by Indian businesses: USD, GBP, EUR, and JPY.

The same business logic is applied consistently across currencies, allowing
comparative exposure analysis without duplicating code.


```python
import pandas as pd

# USD equivalent exposure amount
fx_amount = 10000
it_costs = 600000

# Base exchange rates (Foreign Currency → INR)
base_rates = {
    "USD": 83.0,
    "GBP": 105.0,
    "EUR": 90.0,
    "JPY": 0.56
}

base_rates
```

```python
# Scenario multipliers
scenario_adjustments = {
    "Favorable": -0.02,   # ~2% INR appreciation
    "Neutral": 0.00,
    "Adverse": 0.03       # ~3% INR depreciation
}
```

```python
records = []

for currency, base_rate in base_rates.items():
    for scenario, adj in scenario_adjustments.items():
        scenario_rate = round(base_rate * (1 + adj), 2)
        
        importer_impact = fx_amount * (scenario_rate - base_rate)
        exporter_impact = fx_amount * (base_rate - scenario_rate)
        it_profit = (fx_amount * scenario_rate) - it_costs
        
        records.append([
            currency,
            scenario,
            scenario_rate,
            importer_impact,
            exporter_impact,
            it_profit
        ])

multi_fx_scenarios = pd.DataFrame(
    records,
    columns=[
        "Currency",
        "Scenario",
        "FX_to_INR_Rate",
        "Importer_Impact_INR",
        "Exporter_Impact_INR",
        "IT_Firm_Profit_INR"
    ]
)

multi_fx_scenarios
```

```python
par_records = []

for currency, base_rate in base_rates.items():
    adverse_rate = round(base_rate * 1.03, 2)
    
    par_records.append([
        currency,
        fx_amount * abs(adverse_rate - base_rate)
    ])

profit_at_risk_multi = pd.DataFrame(
    par_records,
    columns=["Currency", "Profit_at_Risk_INR"]
)

profit_at_risk_multi
```

### Break-Even and Pain Thresholds Across Currencies

Break-even remains the base exchange rate for each currency.
Pain thresholds are defined as a ±3% adverse movement.


```python
thresholds_multi = pd.DataFrame({
    "Currency": list(base_rates.keys()),
    "Break_Even_Rate": list(base_rates.values()),
    "Pain_Threshold_Rate": [round(rate * 1.03, 2) for rate in base_rates.values()]
})

thresholds_multi
```

```python
def exposure_zone(rate, base_rate):
    if rate <= base_rate:
        return "Safe"
    elif rate <= base_rate * 1.03:
        return "Warning"
    else:
        return "Danger"

multi_fx_scenarios["Exposure_Zone"] = multi_fx_scenarios.apply(
    lambda x: exposure_zone(x["FX_to_INR_Rate"], base_rates[x["Currency"]]),
    axis=1
)

multi_fx_scenarios
```

## Multi-Currency Interpretation

The extended analysis demonstrates that the exposure modelling framework is
currency-agnostic. Importers face direct cost increases across all currencies
during INR depreciation, while exporters benefit conditionally.

IT services firms exhibit margin sensitivity regardless of currency, with
exposure magnitude varying based on the base exchange rate.

This generalised structure allows seamless integration with downstream
risk scoring and recommendation modules across multiple currencies.
