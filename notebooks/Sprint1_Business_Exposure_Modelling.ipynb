```python
import pandas as pd

# -----------------------------
# 1. Define Business Personas
# -----------------------------

data = {
    "Business Type": ["Importer", "Exporter", "IT Firm"],
    "USD Amount": [10000, 10000, 10000],
    "INR Costs": [None, None, 600000]
}

personas = pd.DataFrame(data)
personas
```

```python
# -----------------------------
# 2. Set Exchange Rates
# -----------------------------

current_rate = 83
rate_1_percent = current_rate * 0.01
rate_5_percent = current_rate * 0.05

rate_1_percent, rate_5_percent
```

```python
# -----------------------------
# 3. Margin Impact Calculation
# -----------------------------

usd_amount = 10000

margin_impact = pd.DataFrame({
    "Scenario": ["1% Change", "5% Change"],
    "Rate Change (₹)": [rate_1_percent, rate_5_percent],
    "Impact on ₹": [usd_amount * rate_1_percent, usd_amount * rate_5_percent]
})

margin_impact
```

```python
# -----------------------------
# 4. Break-even Exchange Rate
# -----------------------------

break_even_data = {
    "Business Type": ["Importer", "Exporter", "IT Firm"],
    "INR Cost / Revenue": [850000, 750000, 600000],
    "USD Amount": [10000, 10000, 10000]
}

break_even_df = pd.DataFrame(break_even_data)
break_even_df["Break-even Rate"] = break_even_df["INR Cost / Revenue"] / break_even_df["USD Amount"]

break_even_df
```

```python
# -----------------------------
# 5. Sensitivity Analysis (₹0.50 movement)
# -----------------------------

rates = [82.5, 83.0, 83.5]

sensitivity = pd.DataFrame({
    "Exchange Rate": rates,
    "Importer Impact (₹)": [(r - current_rate) * usd_amount for r in rates],
    "Exporter Impact (₹)": [(current_rate - r) * usd_amount for r in rates],
    "IT Firm Profit (₹)": [(usd_amount * r) - 600000 for r in rates]
})

sensitivity
```
